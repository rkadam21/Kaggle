```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt, matplotlib.image as mpimg

from scipy import ndimage as ndi
from skimage import feature
```

```python
# Read the training data file
df = pd.read_csv('../Data/train.csv', low_memory = False, header=None, skiprows=1)
```

```python
df_small = df.iloc[:10000]
df_small.head()
```

```python
# extract outputs and inputs to y and x
y = df_small[0]
x = df_small.drop(0, axis=1)
```

```python
def image_prep(row):
    # reshape row
    row = row.values.reshape((28,28))
    
    # invert row
    row = 1 - row
    
    # percentile filter below
    row = abs(ndi.percentile_filter(row, percentile=20, size=2))
#     # Gaussian filter below
#     row = ndi.gaussian_filter(row.values.reshape((28,28)), sigma=1)
    
#     # add sobel figures below
#     sobel = abs(ndi.sobel(row))/100
#     row = row + sobel
    
    return row.reshape(784)
```

```python
x = x.apply(image_prep, axis=1)
x.head()
```

```python
y.head()
```

```python
x.iloc[0, 100:200]
```

```python
# train_test_split?
X_train, X_test, y_train, y_test =  train_test_split(x, y, test_size=0.2, stratify = y)
```

```python
X_test.head()
```

```python
def train(X,Y):
    clf = KNeighborsClassifier(n_neighbors=101, weights='distance')
    clf.fit(X, Y)
    return clf
```

```python
def predict(clf,x,y):
    return (y, clf.predict(x), clf.score(x,y))
```

```python
clf = train(X_train, y_train)
actual, prediction, accuracy = predict(clf, X_test, y_test)
```

```python
accuracy
```

```python
result = pd.DataFrame({'actual':actual, 'prediction':prediction})
result.head(10)
```
