```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import datasets
```

```python
from sklearn.cross_validation import train_test_split
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# Read the training data file
df = pd.read_csv('../Data/train.csv', low_memory = False, header=None, skiprows=1)
df.head()
```

```python
# separarting features from the labels
X = df.drop(0, axis=1)
y = df[0]
```

```python
#splitting dataset in train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#reducing original dataset into a set of 50 concentrated(centers obtained through Kmeans) records per digit.
# The concentrated dataset will have 500 rows, hoping it will be better at predicting the digits with KNeighbors later.
for i in range(0,10):
    digit = df[df[0]==i]
    cluster = KMeans(n_clusters = 50, random_state=0, max_iter = 20).fit(digit)
```

```python
# Just testing out the KMeans algorithm application before running it through the for loop above.
dig = df[df[0]==1]
clus = KMeans(n_clusters = 50, random_state=0, max_iter = 20).fit(dig)
clus
```
